```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import Global as gl
```

```python
def showfunction(df,cols,title,x1=None,x2=None,path=None):
    for i in cols:
        plt.plot(df['t'], df[i], label=i)
    # Set axis labels and a title
    plt.title(title)
    # Set axis labels and a title
    plt.xlabel('Time (seconds)')
    plt.legend()
    if x1:
        plt.xlim(x1, x2)
    # Show the plot
    plt.show()
    if path:
        plt.savefig(path)
        print("saved at ",path)
```

```python
df1_noised = pd.read_csv('GenData/Noised/Full/ToyModel6.csv')
df1 = pd.read_csv('GenData/True/Full/ToyModel6.csv')
```

```python
showfunction(df1_noised,['X0','Xe'],"Noised dynamics evolution",None,None,"Plots/FullModel/Noised/Dynamics_evolution.png")
```

```python
showfunction(df1,['Xe','X0'],"dynamics evolution",None,None,"Plots/FullModel/True/Dynamics_evolution.png")
```

```python
showfunction(df1_noised,['X0','Xe'],"Noised Dynamics evolution",150,190,"Plots/ZoomedModel/Noised/Dynamics_evolution.png")
```
