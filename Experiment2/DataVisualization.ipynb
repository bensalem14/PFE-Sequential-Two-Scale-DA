```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import Global as gl
```

```python
def phase_space(df,cols,x1=None,x2=None,path=None):
    plt.rcParams['agg.path.chunksize'] = 10000  # Increase the chunksize
    plt.rcParams['path.simplify_threshold'] = 0.111111111111  # Adjust the simplify threshold if needed

    plt.plot(df[cols[0]], df[cols[1]])
    plt.xlabel('Position '+x1)
    plt.ylabel('Velocity '+x2)
    plt.title('X Phase Space Representation')
    plt.grid(True)
    plt.show()
```

```python
def showfunction(df,cols,title,x1=None,x2=None,path=None):
    for i in cols:
        plt.plot(df['t'], df[i], label=i)
    # Set axis labels and a title
    plt.title(title)
    # Set axis labels and a title
    plt.xlabel('Time (seconds)')
    plt.legend()
    if x1:
        plt.xlim(x1, x2)
    # Show the plot
    plt.show()
    if path:
        plt.savefig(path)
        print("saved at ",path)
```

```python
df1_noised = pd.read_csv('GenData/Noised/Full/4D_Tokamak.csv')
```

```python
df1=pd.read_csv('GenData/True/Full/4D_Tokamak.csv')
```

```python
phase_space(df1,['Xe1','Ve1'],'Xe1','Ve1')
```

```python
plt.plot(df1['t'], df1['Xe2'],label='Position Xe2')
plt.plot(df1['t'], df1['Ve2'],label='Velocity Ve2')
plt.xlabel('Time t')
plt.title('Y Position/Velocity evolution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
showfunction(df1_noised,['V01','Ve1'],"Noised X_Velocity evolution",None,None,"Plots/FullModel/Noised/X_Velocity_evolution.png")
showfunction(df1_noised,['V02','Ve2'],"Noised Y_Velocity evolution",None,None,"Plots/FullModel/Noised/Y_Velocity_evolution.png")
showfunction(df1_noised,['X01','Xe1'],"Noised X_Position evolution",None,None,"Plots/FullModel/Noised/X_Position_evolution.png")
showfunction(df1_noised,['X02','Xe2'],"Noised Y_Position evolution",None,None,"Plots/FullModel/Noised/Y_Position_evolution.png")
```

```python
showfunction(df1_noised,['Ve1','V01'],"X_Velocity evolution",150,153,"Plots/ZoomedModel/Noised/X_Velocity_evolution.png")
showfunction(df1_noised,['Ve2','V02'],"Y_Velocity evolution",150,153,"Plots/ZoomedModel/Noised/Y_Velocity_evolution.png")
showfunction(df1_noised,['Xe1','X01'],"X_Position evolution",150,153,"Plots/ZoomedModel/Noised/X_Position_evolution.png")
showfunction(df1_noised,['Xe2','X02'],"Y_Position evolution",150,153,"Plots/ZoomedModel/Noised/Y_Position_evolution.png")
```
